# Importing libraries and loading the iris dataset

```python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import tree
```

```python
iris=datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

# Analyzing the data and checking for null values 

```python
df.describe()
```

```python
df.isnull().sum()
```

# Outlier Analysis

```python
df.boxplot(figsize=(12,10))
plt.show()
```

```python
df1 = iris.target
df1
```

# Splitting the datasets into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, df1, test_size=0.25, random_state=40)
```

# Creating Decision Tree Classifier

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_predict = dtc.predict(X_test)
```

# Evaluating Model Performance using Confusion Matrix

```python
confusion_matrix(y_test, y_predict)
```

# Visualizing the Decisiion Tree

```python
Features=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
Class=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 400)

tree.plot_tree(dtc, feature_names = Features, class_names = Class, filled = True);
```
